## **Cleansing**

```
dataset1.loc[dataset1["species"]=="Iris-setosa","species"] = 0
dataset1.loc[dataset1["species"]=="Iris-virginica","species"] = 1
dataset1.loc[dataset1["species"]=="Iris-versicolor","species"] = 2
```

```
dataset1
```

```
dataset1.describe()
```

```
dataset1["species"]=dataset1["species"].astype(float)
```

```
dataset1
```

```
plt.plot(dataset1["sepal_length"], dataset1["petal_length"], "o")
plt.show()
```
